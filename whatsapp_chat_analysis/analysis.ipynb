```python
import re
import pandas as pd
```

```python
f = open(
    r'D:\projects\ML\ML_Shits\whatsapp_chat_analysis\WhatsApp Chat with Dalton Khatri.txt',
    'r',
    encoding='utf-8'
)
```

```python
data = f.read()
```

```python
print(type(data))
```

```python
pattern = r"\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s"
```

```python
messages = re.split(pattern, data)[1:]
len(messages)
```

```python
dates= re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame ({'user_message': messages, 'message_date': dates})

df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %H:%M - ')

df.rename(columns = {'message_date': 'date'}, inplace = True)

df.head()
```

```python
users = []
msgs = []  

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message, maxsplit=1)
    if entry[1:]:  # has username and message
        users.append(entry[1])
        msgs.append(entry[2].strip())
    else:  # system notification without username
        users.append('group_notification')
        msgs.append(entry[0].strip())
```

```python
df['user'] = users
df['message'] = messages
```

```python
df.drop(columns=['user_message'], inplace=True)
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
        if hour == 23:
            period.append(str(hour) + "-" + str('00'))
        elif hour == 0:
            period.append(str('00') + "-" + str(hour + 1))
        else:
            period.append(str(hour) + "-" + str(hour + 1))

         df['period'] = period
    return df
```
