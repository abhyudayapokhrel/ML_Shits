Filtering In Pandas

```python
import pandas as pd
df = pd.read_csv("/data.csv")

#filtering = keeping the rows that match the condition
tall_pokemon = df[df["Height"]>=2]
print(tall_pokemon)
```

Aggregation


```python
#For whole dataFrame

print(df.mean(numeric_only=True))
print(df.sum(numeric_only=True))
print(df.max(numeric_only=True))
print(df.min(numeric_only=True))
print(df.count())
print(df.median(numeric_only=True))
print(df.mode(numeric_only=True))
print(df.std(numeric_only=True))
print(df.var(numeric_only=True))
```

```python
#For single column
print(df["Weight"].mean())
```

Filtering then aggreagation


```python
print(df[df["Height"]>+2]["Weight"].mean())
```

Grouping by same values in coulmn

```python
group = df.groupby("Type1") #group is an object

print(group["Height"].mean()) #this gives mean of each trype as Bug 0.900, Fire 1.21

#this might help to see the bias in the data sampling like peope of one group are asked the questionaire more often than the other:
print(group["Height"].count())
```

Data Cleaning


```python
#Drop columns
df= df.drop(coulmns=["Legendary", "No"])

#2. Handling missing data
#df = df.dropna(subset=["Type2"])  #Every pokemnon missing Type2 will be removed from the dataset
df= df.fillna({"Type2": "None"})

#3. Fix inconsistent values
df["Type1"] = df["type1"].replace({"Grass": "GRASS"},
                                  {"Fire": "FIRE"})

#4. Standerize Text
df["Name"] = df["Name"].str.lower()

#5. Fix DataTypes
df["Legendary"] = df["Legendary"].astype(bool)

#6. Remove Duplicate Values (if any)
df= df.drop_duplicates()

print(df.to_string())
```
