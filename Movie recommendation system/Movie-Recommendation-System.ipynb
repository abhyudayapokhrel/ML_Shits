```python
import numpy as np
import pandas as pd
import ast
import nltk
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(1)
```

Not necessary: original-language as the distribution is skewd towards english movies: movies['original_language'].value_counts()

```python
movies = movies[['movie_id', 'genres', 'title', 'overview', 'keywords', 'cast', 'crew']]
```

```python
movies.head(1)
```

```python
#data preprocessing
movies.isnull().sum()  # shows 3 overviews to be null
movies.dropna() #why does inplace=true doesn't work here? // find out: movies is not original dataset but a slice of another
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# into ['Action', 'Adventure', 'Fantasy']

def convert(obj):
  L =[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L

  #the problem is we do not get lsit of dicts but a string of taht so convert string od list of dict to list of dict
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#for cast : top 3 cast
def convert3(obj):
  L =[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter += 1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.dropna(inplace=True)
movies['overview'].isnull().sum()
```

```python
#convert overwiew from string to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y =[]

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
# swastidonhoma mero nam swastidon ho
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#find similarity: cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors) #saving the similarity of 4806 movies with 4806 movies
```

Main function of recommending

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6] #key is for sorting not on the basis of index but of similarity distance, reverse so that most similar would be on top(descending), enumeratre s used here to keep track of both the similarity score and the corresponding movieâ€™s index in new_df.

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('The Godfather')
```
